# Merger Device Working Principle

Vuong and Anna (2012) *Biomicrofluidics* presents geometric model for bubbles in rectangular channels: http://www.ncbi.nlm.nih.gov/pmc/articles/PMC3360712/

The alkaline lysis device has five liquid inlets with adjustable flow rates ($q_n$): cells ($q_1$), alkaline lysis buffer ($q_2$), oil for cell/alk ($q_3$), PCR mix ($q_4$), oil for PCR mix ($q_5$). By varying these flow rates one can control the drop making, pairing and merging. 

This Notebook describes the basics of controlling drop sizes and pairing.

## Drop size is a function of nozzle size and flow rate ratio

Assume $h \leq w$. 
- Pancake shape when $h \leq D \leq w$:
$$ V_{pancake} = \frac{\pi h^3}{6} + \frac{\pi h}{4}(D-h) \big(\frac{\pi h}{2}+D-h \big)$$

- Plug/slug shape when $D \geq w$:
$$ V_{slug} = V_{box} + V_{pancake,max} = (D-w)wh + \big[\frac{\pi h^3}{6}+\frac{\pi h}{4} (w-h) \big( \frac{\pi h}{2}+w-h \big) \big] $$

The size of drop at coflow cell encapsulation junction is mostly governed by the nozzle size ($w$) and the ratio between aqueous and oil phase flow rates ($Q_{aq}/Q_{oil}$) when drops are generated by the 'plugging' mechanism (assuming smallest dimension serves as the orifice):

$$\frac{D}{h} = 1 + \alpha \frac{Q_{aq}}{Q_{oil}}$$

For simplicity, let's assume $\alpha = 1$.

## Coflow junction for cell encapsulation & lysis

If $q_1 = q_2 = 50$ µL/hr and $h = 30$ µm, $w = 40$ µm (that is, $Q_{aq} = 100$ and $h^2 w = 36$ pL), then

- Rectangular approximation:
$$ V_{rect} = D w h = 36 \big( 1 + \frac{100}{q_3} \big) \textrm{[pL]}$$
$$ \nu_{rect} = \frac{q_1+q_2}{V_r}=\frac{770 q_3}{q_3 + 100} \textrm{[drops/sec]} $$


- More accurately, the pancake/slug approximation can be used:

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# channel dimension
h = 30.0; w = 40.0; Q1 = 100.0

D = lambda x: h*(1+Q1/x)
# rectangle
v_rect = lambda x: D(x)*h*w
# pancake
v_panc = lambda x: np.pi*h**3/6 + np.pi*h/4*(D(x)-h)*(np.pi*h/2+D(x)-h)
# slug
v_slug = lambda x: np.pi*h**3/6 + np.pi*h/4*(w-h)*(np.pi*h/2+w-h)+w*h*(D(x)-w)
# drop diameter
dd = lambda x: (x/(4*np.pi/3))**(1.0/3.0)*2

x = np.linspace(Q1*0.5, Q1*10, num=50)
# pancake-like regime
x1 = np.linspace(Q1/(w/h-1),Q1*10, num=25)
# slug-like regime
x2 = np.linspace(Q1*0.5, Q1/(w/h-1), num=25)

print "1:2 ratio makes", dd(v_slug(200)), "um Cell drops."

ax = plt.subplot(121)
plt.plot(x, dd(v_rect(x)), 'k--', label="rect")
plt.plot(x1, dd(v_panc(x1)), 'r', label="panc")
plt.plot(x2, dd(v_slug(x2)), 'b', label="slug")
plt.plot(x2, dd(v_panc(x2)), 'r--')
#plt.plot(x2, y1x2)
plt.title('Diameter vs. $q_3$')
plt.ylabel('Drop Diameter [um]')
plt.xlabel('Cell Oil Flow Rate [uL/hr]')
legend = ax.legend()
frame = legend.get_frame()
frame.set_facecolor('white')
ax.grid(ls=':')
#ax.set_xticks(list(range(0,1100,200)))

ax = plt.subplot(122)
plt.plot(x, Q1*10**9/v_rect(x)/3600, 'k--')
plt.plot(x1, Q1*10**9/v_panc(x1)/3600, 'r', label="panc")
plt.plot(x2, Q1*10**9/v_slug(x2)/3600, 'b', label="slug")
plt.title('Freq. vs. $q_3$')
plt.ylabel('Frequency [drops/sec]')
plt.xlabel('Cell Oil Flow Rate [uL/hr]')
ax.yaxis.tick_right()
ax.grid(ls=':')
ax.yaxis.set_label_position("right");
```

## PCR drop making junction

Continuing with the same scaling law for drop formation, 

If $q_4 = 1000$ µL/hr and $w = 60$ µm, $h = 80$ µm, then
(*Note: Actual design has 40 µm neck before 80 µm nozzle*)

- Rectangular approximation:
$$ V_{\textrm{PCR}} = 288 \big( 1 + \frac{1000}{q_5} \big) \textrm{[pL]}$$
$$ \nu_{\textrm{PCR}} = \frac{964 q_5}{q_5 + 1000} \textrm{[drops/sec]} $$


- *The plots below used discoid approximation.*

```python
# channel dimension
h = 60.0; w = 80.0; Q1 = 1000.0

D = lambda x: h*(1+Q1/x)
# rectangle
v_rect = lambda x: D(x)*h*w
# pancake
v_panc = lambda x: np.pi*h**3/6 + np.pi*h/4*(D(x)-h)*(np.pi*h/2+D(x)-h)
# slug
v_slug = lambda x: np.pi*h**3/6 + np.pi*h/4*(w-h)*(np.pi*h/2+w-h)+w*h*(D(x)-w)
# drop diameter
dd = lambda x: (x/(4*np.pi/3))**(1.0/3.0)*2

x = np.linspace(Q1*0.5, Q1*5, num=50)
# pancake-like regime
x1 = np.linspace(Q1/(w/h-1),Q1*5, num=25)
# slug-like regime
x2 = np.linspace(Q1*0.5, Q1/(w/h-1), num=25)

print "1:1 ratio makes", dd(v_slug(1000)), "um PCR drops."

ax = plt.subplot(121)
plt.plot(x, dd(v_rect(x)), 'k--', label="rect")
plt.plot(x1, dd(v_panc(x1)), 'r', label="panc")
plt.plot(x2, dd(v_slug(x2)), 'b', label="slug")
plt.plot(x2, dd(v_panc(x2)), 'r--')
#plt.plot(x2, y1x2)
plt.title('Diameter vs. $q_3$')
plt.ylabel('Drop Diameter [um]')
plt.xlabel('Cell Oil Flow Rate [uL/hr]')
legend = ax.legend()
frame = legend.get_frame()
frame.set_facecolor('white')
ax.grid(ls=':')
ax.set_xticks(list(range(0,int(Q1)*6,1000)))

ax = plt.subplot(122)
plt.plot(x, Q1*10**9/v_rect(x)/3600, 'k--')
plt.plot(x1, Q1*10**9/v_panc(x1)/3600, 'r', label="panc")
plt.plot(x2, Q1*10**9/v_slug(x2)/3600, 'b', label="slug")
plt.title('Freq. vs. $q_3$')
plt.ylabel('Frequency [drops/sec]')
plt.xlabel('Cell Oil Flow Rate [uL/hr]')
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")
ax.grid(ls=':')
ax.set_xticks(list(range(0,int(Q1)*6,1000)));
```

## Pairing Ratio

40.0 um cell drops can be made at 1:2 ratio. 92.4 um PCR drops can be made at 1:1 ratio.

If flow rates are fixed at such ratios, i.e. $(q_1+q_2)/q_3 = 0.5$ and $q_4/q_5 = 1$, the drop sizes are kept constant and the drop generation frequency depends solely on the aqueous phase flow rates. In this case, we can control the pairing ratio by controlling the ratio between cell influx ($q_1+q_2$) and PCR influx ($q_4$).

Ideally, 1:1 pairing ratio can be achieved when
$$ \frac{q_4}{q_1+q_2} = \big( \frac{92.4}{40.0} \big)^3 = 12.3 $$
as predicted by the drop size ratio.

```python
# channel dimension
h1 = 30.0; w1 = 40.0; q1 = 100.0
h2 = 60.0; w2 = 80.0; q2 = 1000.0

D_gen = lambda x, h, q1: h*(1+q1/x)
# slug
v_slug_opt = lambda x, h, w, q1: np.pi*h**3/6 + np.pi*h/4*(w-h)*(np.pi*h/2+w-h)+w*h*(D_gen(x,h,q1)-w)

x = np.linspace(1,25, num=50)

ax = plt.plot(x, (q1*10**9/v_slug_opt(q1*2, h1, w1, q1)/3600)/(q1*x*10**9/v_slug_opt(q1*x, h2, w2, q1*x)/3600))
plt.title('Pairing Ratio: # Cell drops / # PCR drops')
#plt.ylabel('Frequency [drops/sec]')
plt.xlabel('$q_4/(q_1+q_2)$', fontsize=20)
plt.ylim([0,6])
plt.grid(ls='--',which='both')
```
